```python
import pennylane as qml
from pennylane import numpy as np
```

```python
n_qubits = 1

device = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(device)
def qc(angle):
    qml.RY(angle,wires=0)

    return qml.probs(wires=0)
```

```python
qc(2*np.arcsin(np.sqrt(0.3)))
```

```python
n_qubits = 2

device = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(device)
def qc():
    qml.Hadamard(wires=0)

    return qml.state()
```

```python
qc()
```

```python
qml.draw_mpl(qc, show_all_wires=True)()
```

```python
n_qubits = 20

device = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(device)
def qc():

    qml.PauliX(wires=0)
    qml.PauliZ(wires=0)

    return qml.state()
```

```python
qc()
```

```python
n_qubits = 20

device = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(device)
def qc():

    for i in range(n_qubits):
        qml.PauliX(wires=i)

    return qml.state()
```

```python
qml.draw_mpl(qc, show_all_wires=True)()
```

```python
n_qubits = 20

device = qml.device("default.qubit", wires=n_qubits, shots=1000)

@qml.qnode(device)
def qc():

    for i in range(n_qubits):
        qml.Hadamard(wires=i)

    return qml.probs()
```

```python
qml.draw_mpl(qc)()
```

```python
n_qubits = 5

device = qml.device("lightning.qubit", wires=n_qubits, shots=1000)

@qml.qnode(device)
def qc(angles):

    for i in range(n_qubits):
        qml.RY(angles[i], wires=i)

    return qml.probs()
```

```python
from pennylane import numpy as np
```

```python
params = np.random.random(n_qubits)



qc(params)
```
